## Ploting The Closing Price Of The Company's Stock.

```python
plt.figure(figsize=(20,12))
for i,company in enumerate(tech_list,1):
    plt.subplot(2,2,i)
    df=full_df[full_df['Name']==company]
    color=['yellow','red','blue','green','purple']
    list_c=['list','Apple','Google','Microsoft','Amazon']
    plt.plot(df['date'],df['close'],color=color[i])
    plt.xticks(rotation='vertical')
    plt.title(list_c[i],color=color[i],fontsize=16)
    plt.tight_layout()
```

```python
## Conclusion - All the above stocks have an upward trend over the years 
```

```python
## Conclusion : The std of 1 means 68% of the data lies in this region. STD of 2 means 95% of data lies in this region
## and std of 3 means 99% of data lies in this region.
                
```
